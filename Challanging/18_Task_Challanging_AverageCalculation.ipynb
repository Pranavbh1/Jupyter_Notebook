    def average_grade():
        try:
            total_std = (input("Enter total number of students:  "))
            if total_std.isalpha() or int(total_std)<=0 :
                raise ValueError('NegativeValueError : Total student must be more than 0')

            totalgrade = 0
            for _ in range(0,int(total_std)):
                name_std = input("Enter the name of student: ")
                sub1 = (input("Enter grade of subject 1, out of 100:  "))
                sub2 = (input("Enter grade of subject 2, out of 100:  "))
                sub3 = (input("Enter grade of subject 3, out of 100:  "))

                if sub1=='' and sub2=='' and sub3=='':
                    raise ZeroDivisionError
                if not (1 <= int(sub1) <= 100 and 1 <= int(sub2) <= 100 and 1 <= int(sub3) <= 100):
                    raise ValueError(" ValueError : Grades should be within the range 1-100")



                avg_ = [int(sub1), int(sub2), int(sub3)]
                avg_grade = sum(avg_)/ len(avg_)
                print(f"\nAverage grade for : {name_std} is {avg_grade}\n")
                totalgrade += avg_grade

            print(f"\nAverage grade for all students is {totalgrade / int(total_std)}\n")

        except ValueError as ve:
            print(ve)
        except ZeroDivisionError:
            print(" ZeroDivisionError : It will return in raising the problem in formulating average of particular student.")
            
        finally:
            print("Average Calculation Done Successfully")
