if -elif - else

In Python, if, elif, and else are conditional statements used to control the flow of execution based on certain conditions. They are part of the control flow statements and allow you to create branching logic in your code.

* .The if statement checks a condition. If the condition is true, the code block associated with the if statement is executed.

* .If the condition in the if statement is false, the program moves on to the next elif statement (optional). It checks its condition and executes the associated code block if the condition is true.

* . You can have multiple elif statements in order to test multiple conditions. The program will move to the next elif statement only if the previous conditions were false.

* .If none of the conditions in the if and elif statements are true, the else block (optional) is executed. It acts as a catch-all and is executed when all preceding conditions are false.
